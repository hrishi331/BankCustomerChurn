```python
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
import pickle
```

```python
df = pd.read_csv('Customer-Churn-Records.csv')
df.head()
```

```python
X_train,X_test,y_train,y_test = train_test_split(df.drop('Exited',axis=1),df['Exited'],test_size=0.2,random_state=42)
```

```python
X_test.head()
```

```python
y_test.head()
```

```python
# pickle 1
def col_dropper(df):
  return df.drop(['RowNumber','CustomerId','Surname'],axis=1)

X_train = col_dropper(X_train)
X_test = col_dropper(X_test)
```

```python
train_df = pd.concat([X_train,y_train],axis=1)
test_df = pd.concat([X_test,y_test],axis=1)
```

```python
test_df
```

```python
with open("train_df.pkl","wb") as file:
  pickle.dump(train_df,file)
with open("test_df.pkl","wb") as file:
  pickle.dump(test_df,file)
```
